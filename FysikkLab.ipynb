##3 Teori


I begynnelsen av eksperimentet blir viktige begreper innført, som energibevaring og treghetsmoment for kula. En sentral parameter er kulas treghetsmoment, $(I)$, som avhenger av masse $(m)$ og radius $(r)$. For enklere beregninger antas kula å ha en jevn massefordeling, med $c = 2/5$ som en viktig konstant.
Treghetsmomentet er så gitt ved:<br>

<center> $I = cmr^2$ (Likning 1)</center>



Kula starter fra ro og ruller deretter nedover en krum bane med en kjent form $y(x)$. Prinsippet om energibevaring gir en sammenheng mellom kulas hastighet $(v)$ og dens høyde $(y)$ langs banen. Dette tillater oss å beregne hastigheten som en funksjon av både høyden og den horisontale posisjonen $(x)$.
<center>$v(y) = \sqrt{\frac{2g(y_0-y)}{1+c}}$ (Likning 2)</center>


Helningsvinkelen er avgjørende for å forstå banens form og hvordan den påvirker kulas bevegelse. Den gir et mål på hvor mye banen buer og hvor bratt den er på et spesifikt punkt.
Den beregnes ved å ta arctan av stigningstallet, som er forholdet mellom endringen i vertikal høyde $(dy)$ og endringen i horisontal avstand $(dx)$ på banen, som vil være det samme som den deriverte av farten.

Matematisk uttrykt:

<center>$β = arctan(\frac{dy}{dx})$ (Likning 3)</center>




Banens krumning, representert som den inverse krumningsradien κ, bestemmer sentripetalakselerasjonen $(a_⊥)$. Positiv krumning fører til en positiv $(a_⊥)$, og negativ krumning fører til en negativ $(a_⊥)$.

Den inverse krumninksradien er gitt ved:

<center>$κ = \frac{y′′}{[1 + (y′)^2]^{3/2}}$ (Likning 4)</center>

Sentripetalakselølerasjonen er så gitt ved:

<center>$a_⊥ = v^2/R = v^2κ =\frac{2g(y_0 − y)}{1 + c}·\frac{y′′}{[1 + (y′)^2]^{3/2}}$ (Likning 5)</center>


Friksjonskraften er den kraften som virker mellom kula og banen, og den motvirker rullende bevegelse. Friksjonskraften er viktig for å opprettholde ren rulling uten gliding. For kula beregnes friksjonskraften slik:

<center>$f = \frac{c}{1+c}mg*sin(β)$ (Likning 6)</center>




Normalkraften $N$ som virker normalt på banen, bestemmes ved å bruke Newtons andre lov. $N$ er avhengig av tyngdekraften $(mg)$ og sentripetalaksellerasjonen $(a⊥)$. Det er viktig å merke seg at $N$ kan være større eller mindre enn $mg$ avhengig av hvordan banen krummer.

Normalkraften er gitt ved:

<center>$N = m(g*cos(β) + a_⊥)$ (Likning 7)</center>


For å sikre at kula ruller rent uten å gli på banen, må ikke friksjonskraften $(f)$ overskride sin egne maksimale verdi $(µ_s|N|)$. For en gummibelagt ball antas friksjonskoeffesienten $(µ_s)$ til å være $0,4$. Hvis $f/N$ overstiger denne koeffisienten , vil kula begynne å gli, og ren rulling blir umulig.

<center>$\frac{|f|}{|N|} \leq µ_s$ (Likning 8)</center>


Labprosjektet presenterer to metoder for å beregne tidsutviklingen for kula som ruller på banen. Den første metoden bruker Newtons 2. lov og Euler-metoden med et fast tidssteg $(dt)$. Den andre metoden benytter energibevaring og et fast horisontalt steg $(dx)$, hvor tiden $(dt)$ beregnes som $dt = dx/vx$. Denne rapporten tar for seg metode 2.

Metode 2 baserer seg på energibevaring og bruker endringer i horisontal forflytning $(dx)$ for å forstå kulas tidsutvikling. Ved å beregne gjennomsnittlig horisontal hastighet $(vx)$ mellom to påfølgende posisjoner $(xn, yn)$, bestemmes tidsintervallet $(∆tn)$ som kula bruker på den horisontale forskyvningen $(∆xn)$. Dette gjentas langs hele banen, og dermed beregnes tidsutviklingen. Metode 2 gir en enklere tilnærming ved å forstå kulas bevegelse gjennom horisontal forflytning og hastighet, og den tillater beregning av tiden det tar for kula å rulle fra start til mål på den krumme banen. I dette forsøket er denne metoden valgt til å beregne tidsutviklingen.


<center>$t_n =\sum_{J=1}^{n}∆tj ; n = 1, . . . , 1400$ (Likning 9)</center>

```python
c = 2/5
m = 31
r = 11
g = 9810
I_0 = c * m * r**2
def v(x):
  return ((10*g*(cs(0)-cs(x))/7))**(1/2)

def derivert(x):
  return cs(x, 1)

v_exp = data[:, 3] * 1000
plt.figure(figsize=(10, 5))
plt.title('Fart', fontsize=20)
plt.xlabel('$x$ (mm)',fontsize=20)
plt.ylabel('v(x) (mm/s)',fontsize=20)

numerical, = plt.plot(x, v(x), 'r-')
experimental, = plt.plot(x_exp, v_exp, 'b-')
plt.legend([numerical, experimental], ['Ballens beregnede fart', 'Ballens målte fart'], loc='best')

plt.axhline(0, color='black')
plt.grid()
plt.show()
```

**Beregnet helningsvinkel $h(x)$**

```python
def h(x):
  return np.tan(derivert(x))

plt.figure(figsize=(10, 5))
plt.plot(x,h(x)*180/np.pi)
plt.title('Helningsvinkel', fontsize=20)
plt.xlabel('$x$ (mm)',fontsize=20)
plt.ylabel('$h(x)$ (⁰)',fontsize=20)
plt.ylim(-30,30)
plt.xlim(-50,1450)
plt.axhline(0, color='black')
plt.grid()
plt.show()
```

**Beregnet friksjonskraft $f(x)$**

```python
def F(x):
  return 2*m*g*10**(-6)*np.sin(h(x))/7

plt.figure(figsize=(10, 5))
plt.plot(x, F(x))
plt.title('Friksjonskraft', fontsize=20)
plt.xlabel('$x$ (mm)',fontsize=20)
plt.ylabel('$F(x)$ (F)',fontsize=20)
plt.ylim(-0.05,0.05)
plt.xlim(-50,1450)
plt.axhline(0, color='black')
plt.grid()
plt.show()
```

Figur 4: Friksjonskraften på ballen langs banen

Friksjonskraften på ballen som beveger seg langs banen vises på figur 4. Der kan man se at absoluttverdien til friksjonskraften er størst på ca. $300mm$ og ca. $700mm$, og friksjonsgrafen har vendepunkt på ca. $500mm$ og ca. $900mm$.

**Beregnet normalkraft $N(x)$**

```python
def a_s(x):
  return 2*g*(cs(0)-cs(x))/(1+c)*(cs(x,2)/(1+cs(x,1)**2)**(3/2))
def N(x):
  return m*(g*np.cos(h(x))+a_s(x))*(10**(-6))
plt.figure(figsize=(10, 5))
plt.plot(x, N(x))
plt.title('NormalKraft', fontsize=20)
plt.xlabel('$x$ (mm)',fontsize=20)
plt.ylabel('$N(x)$ (N)',fontsize=20)
plt.ylim(0,1)
plt.xlim(-50,1450)
plt.axhline(0, color='black')
plt.grid()
plt.show()
```

Figur 5: Normalkraften på ballen langs banen

Toppunktet i banen "Figur 1" sammenfaller med bunnpunktet i normalkraftgrafene "Figur 5". Ved toppunktet av banen er kula nær toppen av banen og
opplever minimal kontakt med underlaget, og derfor er normalkraften mindre. Bunnpunktet i banen sammenfaller med toppunktet i
normalkraftgrafene fordi normalkraften må være større enn tyngdekraften for at kula skal kunne opprettholde sin bane. Ved bunnpunktet av
banen, der kula beveger seg oppover og endrer retning, må normalkraften være tilstrekkelig stor for å motvirke tyngdekraften og opprettholde
kulas bane.

**Beregnet forhold $|f(x)/N(x)|$**

```python
def F_N(x):
  return (abs(F(x)/N(x)))

plt.figure(figsize=(10, 5))
plt.plot(x, F_N(x))
plt.title('f/N', fontsize=20)
plt.xlabel('$x$ (mm)',fontsize=20)
plt.ylabel('$f/N$',fontsize=20)
plt.ylim(0,0.2)
plt.xlim(0,1450)
plt.axhline(0, color='black')
plt.grid()
plt.show()
```

Figur 6: Forhold mellom friksjonskraften og normalkraften



```python
# Printe Max verdi av F(x)/N(x)

x_values = np.arange(0, 1401)
y_values = F_N(x_values)
max_value = np.max(y_values)
max_index = np.argmax(y_values)
x_at_max = x_values[max_index]

print(f"Maximum value: {max_value} at x = {x_at_max}")
```

På figur 6 ser man at sammenhengen mellom friksjonskraft og normalkraft gjenspeiler ballens oppførsel langs banen "Figur 1". Når $f/N = 0$ er det ingen friksjonskraft, og det skjer på ca. $500mm$ og ca. $900mm$, som er på toppunktet og bunnpunktet til banen. Når $f/N$ er størst blir det også  lav normalkraft "Figur 5", som kommer fram på $264mm$ og ca. $750mm$ hvor $f/N$ er høy og normalkraften er lav. Som vist i kodeblokken over er $f/N = 0,14$ når den har størst verdi

**Tidsutvikling $t(x)$**

```python
t_exp = data[:, 0]

def v_x(x):
  return (v(x)*np.cos(h(x)))

def dt(n):
  return 2/(v_x(n-1)+v_x(n))

def t(n1):
    t = 0
    n = 1
    while(n1>=n):
      t+=dt(n)
      n += 1
    return t

t_arr = []
n=0
for i in range(1400):
  t_arr.append(t(n))
  n=n+1

x_2 = np.linspace(0,1400,1400)

plt.figure(figsize=(10, 5))
plt.title('tid', fontsize=20)
plt.xlabel('$x$ (mm)',fontsize=20)
plt.ylabel('$t(x)$ (s)',fontsize=20)
plt.axhline(0, color='black')
plt.grid()
numerical, = plt.plot(x_2, t_arr, 'r-')
experimental, = plt.plot(x_exp, t_exp, 'b-')

plt.legend([numerical, experimental], ['Beregnet tidsutvikling', 'Målt tidsutvikling'], loc='best')

plt.show()
```
